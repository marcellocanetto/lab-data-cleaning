Bonus Question: Create a dataframe with the outliers you have identified in the dataframe and export it to a csv file in your-code folder.

```python
stats = data5.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=data5.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = data5[(data5[col] < lower) | 
                   (data5[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```
